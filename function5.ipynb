Q6.Explain the concept of generators in Python and how they are defined

 Ans: ##Generator
  * generator are a special type of iterator that allow you to iterate over a sequence of values
  * it uses yeild  keyword instead of return .
  * Generators can be defined in two main ways:

Generator Functions:

Generator expression:

```python
# Generator Functions:
def countdown(n):
    while n > 0:
        yield n
        n =n-1

# Using
for number in countdown(5):
    print(number)
```

Q10.0. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?





ANS : map()

Syntax: map(function, iterable, ...)

Example:
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))  # Outputs: [1, 4, 9, 16]

2. reduce()
Purpose: reduces the iterable to a single cumulative value. It is part of the functools module.

Syntax: reduce(function, iterable[, initializer])
*
from functools import reduce

numbers = [1, 2, 3, 4]
sum_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_numbers)
  # Outputs: 10

3. filter()
Purpose: Applies a given function to all items in an iterable and returns an iterator yielding those items for which the function returns True.

Syntax: filter(function, iterable)

Example:
numbers = [1, 2, 3, 4]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
